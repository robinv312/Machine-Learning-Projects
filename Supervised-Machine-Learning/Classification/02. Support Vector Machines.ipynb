## Introduction to Support Vector Machines

* SVM as it is one of the powerful technique for Classification, Regression  & Outlier detection with an intuitive model.
* For a dataset consisting of features set and labels set, an SVM classifier builds a model to predict classes for new examples. It assigns new example/data points to one of the classes. If there are only 2 classes then it can be called as a Binary SVM Classifier.
* There are Two types of Support Vector Machines i.e., Linear and Non Linear Support Vector Machines depending upon the Kernels chosen for the task of Classification of Targets.
* SVM is a classifier formally defined by a separating hyperplane. An hyperplane is a subspace of one dimension less than its ambient space. 



Therefore the hyperplane of a two dimensional space above is a one dimensional line dividing the red and blue dots.

## Why do we need Kernels ?



* In machine learning, a “kernel” is usually used to refer to the kernel trick, a method of using a linear classifier to solve a non-linear problem. 

* It entails transforming linearly inseparable data like to linearly separable ones.

* The kernel function is what is applied on each data instance to map the original non-linear observations into a higher-dimensional space in which they become separable.

####  Types of Kernels :

* **Linear Kernel**:  This Kernel is simply used for Solving Linearly Separable linear Problems.
* **Polynomial Kernel**: This Kernel is mainly used for Image Processing.


* **Gaussian Kernel**: This Kernel is used for general purposes, like in case when we have no idea about the data, we have.


* **RBF Kernel**: It is one of the most popular kernels. For distance metric squared euclidean distance is used here. It is used to draw completely non-linear hyperplanes.

where x & x’ are vectors of feature space. \sigma  is a free parameter. Selection of parameters is a critical choice. Using a typical value of the parameter can lead to overfitting our data.

* **Hyperbolic Tangent Kernel and Sigmoid Kernel**: It is generally used for neural networks
* **ANOVA RBF Kernel**: It can be used for Regression Problems.
* **Linear Splines Kernel**: It is useful when dealing with large sparse data vectors. It is often used in text categorization. The splines kernel also performs well in regression problems.


## Selecting the Hyperplanes

* **For Linearly Separable**: 
    * For the data which can be separated linearly, we select two parallel hyperplanes that separate the two classes of data,  so that distance between both the lines is maximum.
    * The region b/w these two hyperplanes is known as “margin” & maximum margin hyperplane is the one that lies in the middle of them.
   





* **For Non Linearly Separable**: To build classifier for non-linear data, we try to minimize


    * Here, max() method will be zero( 0 ), if xi is on the correct side of the margin. For data that is on opposite side of the margin, the function’s value is proportional to the distance from the margin.

    * where, lambda  determines tradeoff b/w increasing the margin size and that xi is on correct side of the margin.

## Adavntages of SVM

* SVMs are effective when the number of features is quite large.

* It works effectively even if the number of features are greater than the number of samples.

* Non-Linear data can also be classified using customized hyperplanes built by using kernel trick.

* It is a robust model to solve prediction problems since it maximizes margin.

* Uses a subset of training points in the decision function (called support vectors), so it is also memory efficient.

* 


## Disadvantages of SVM

* The biggest limitation of Support Vector Machine is the choice of the kernel. The wrong choice of the kernel can lead to an increase in error percentage.

* With a greater number of samples, it starts giving poor performances.

* SVMs have good generalization performance but they can be extremely slow in the test phase.

* SVMs have high algorithmic complexity and extensive memory requirements due to the use of quadratic programming.


**Note : In Python we have Packages such as SVC, NuSVC, and LinearSVC**

* The Only Difference between **SVC** and **NuSVC** is that **SVC** uses **```C```** Parameter whereas **NuSVC** uses **```nu```** Parameter. Where **```nu```** is upper bound for no.of error. lower bound for no.of support vectors.

* **LinearSVC** is an SVC for Classification that uses only linear kernel. In LinearSVC(), we don’t pass value of kernel, since it’s specifically for linear classification.

## Parameters 

* **C**: float, optional, default = 1.0, 
    * Regularization parameter. The strength of the regularization is inversely proportional to C. Must be strictly positive. The penalty is a squared l2 penalty.

* **Kernel**: string, optional, default = 'rbf'
    * Specifies the kernel type to be used in the algorithm. It must be one of ‘linear’, ‘poly’, ‘rbf’, ‘sigmoid’, ‘precomputed’ or a callable. If none is given, ‘rbf’ will be used. If a callable is given it is used to pre-compute the kernel matrix from data matrices; that matrix should be an array of shape

* **degree**: int, optional, default = 3
    * Degree of the polynomial kernel function (‘poly’). Ignored by all other kernels.
    
* **gamma**: {'scale' or 'auto'} or float, optional, default = 'scale'
    * Kernel coefficient for ‘rbf’, ‘poly’ and ‘sigmoid’.
    * if gamma='scale' (default) is passed then it uses 1 / (n_features * X.var()) as value of gamma,
    * if ‘auto’, uses 1 / n_features.
    
* **coef0**: float, optional, default = True
    * Independent term in kernel function, It is only significant in poly and sigmoid.
    
* **shrinkage**: boolean, optional, default = True
    * whether to use the Shrinking heuristic.
    
* **probability**: boolean, optional, default = False
    * Whether to enable probability estimates. This must be enabled prior to calling fit, will slow down that method as it internally uses 5-fold cross-validation, and predict_proba may be inconsistent with predict.
   
* **tol**: float, optional, default = 1e-3
    * Tolerance for stopping criterion.

* **cache_size**: float, optional
    * Specify the size of the kernel cache (in MB).
   
* **class_weight**: {dict, 'balanced'}, optional
    * Set the parameter C of class i to class_weight[i]*C for SVC. If not given, all classes are supposed to have weight one. The “balanced” mode uses the values of y to automatically adjust weights inversely proportional to class frequencies in the input data as n_samples / (n_classes * np.bincount(y))
    
* **verbose**: bool, default = False
    * Enable verbose output. Note that this setting takes advantage of a per-process runtime setting in libsvm that, if enabled, may not work properly in a multithreaded context.
   
* **max_iter**: int, optional, default = False
    * Hard limit on iterations within solver, or -1 for no limit.
    
* **decision_function_shape**: ‘ovo’, ‘ovr’, default=’ovr’
    * Whether to return a one-vs-rest (‘ovr’) decision function of shape (n_samples, n_classes) as all other classifiers, or the original one-vs-one (‘ovo’) decision function of libsvm which has shape (n_samples, n_classes * (n_classes - 1) / 2). 
    * However, one-vs-one (‘ovo’) is always used as multi-class strategy.

* **break_ties**: bool, optional, default = False
    * If true, decision_function_shape='ovr', and number of classes > 2, predict will break ties according to the confidence values of decision_function; otherwise the first class among the tied classes is returned. Please note that breaking ties comes at a relatively high computational cost compared to a simple predict.

* **random_state**: int, RandomStateInstance or None, default = False
    * The seed of the pseudo random number generator used when shuffling the data for probability estimates. If int, random_state is the seed used by the random number generator; If RandomState instance, random_state is the random number generator; If None, the random number generator is the RandomState instance used by np.random.

````python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
````

## Scenario - 1 

*  Using a simple dataset, to check the working functionality

````python
# lets import the data
data = pd.read_csv('iris/iris.csv')

# lets check the shape of the data
data.shape
````

````python
# lets check the head of the data
data.head()
````

````python
# lets split the target and features

y = data['Species']
x = data.drop(['Id','Species'], axis = 1)
````

````python
# train test split

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
````

````python
# standardization

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
````

````python
# modelling

from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report

model = SVC()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot = True)
cr = classification_report(y_test, y_pred)
print(cr)
````

## Scenario - 2

* Using a Bigger Dataset, to check the Functionality, with Parameters and also tuning the Parameters using Grid search and Randomized Search algorithms.

````python
# lets import the dataset
data = pd.read_csv('cancer/data.csv')

# lets check the shape of the dataset
data.shape
````

````python
# lets check the head of the dataset
pd.set_option('max_columns', 35)
data.head()
````

````python
# features and target splitting

y = data['diagnosis']
x = data.drop(['id', 'Unnamed: 32','diagnosis'], axis = 1)
````

````python
# train test split

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
````

````python
# standardization
import warnings
warnings.filterwarnings('ignore')

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
````

````python
# modelling

from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report

model = SVC()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot = True)
cr = classification_report(y_test, y_pred)
print(cr)
````

````python
# getting the real time values

prediction = model.predict(sc.transform(np.array([[11, 18, 125,1585, 0.252, 0.523,0.085,0.07,0.1856, 0.05995, 
                                                   1.096, 1.52, 9.6, 158.6, 0.00056, 0.2,
                    0.0014,0.0025,0.002, 0.145,28.6, 23.3,145.2, 1002,0.0025,0.02541, 0.145,0.0021,0.001, 0.002 ]])))
print("Prediction :", prediction)
````

## Hyper Parameter Tuning

* In machine learning, hyperparameter optimization or tuning is the problem of choosing a set of optimal hyperparameters for a learning algorithm. A hyperparameter is a parameter whose value is used to control the learning process. By contrast, the values of other parameters (typically node weights) are learned.
