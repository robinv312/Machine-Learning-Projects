# THE MOVIE REVIEW PROJECT ON NLP

```python
import numpy as np
import pandas as pd
import os
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
print(os.listdir(nltk.data.find('corpora')))
```

```python
from nltk.corpus import movie_reviews
```

```python
print(movie_reviews.categories())
```

```python
print(len(movie_reviews.fileids('pos')))
print(' ')
print(movie_reviews.fileids('pos'))
```

```python
neg_rev = movie_reviews.fileids('neg')
len(neg_rev)
neg_rev
```

```python
rev = nltk.corpus.movie_reviews.words('pos/cv000_29590.txt')
```

```python
rev
```

```python
rev_list = []
```

```python
for rev in neg_rev:
    rev_text_neg = rev = nltk.corpus.movie_reviews.words(rev)
    review_one_string = " ".join(rev_text_neg)
    review_one_string = review_one_string.replace(' ,',',')
    review_one_string = review_one_string.replace(' .','.')
    review_one_string = review_one_string.replace("\' " ,"'")
    review_one_string = review_one_string.replace(" \'" ,"'")
    rev_list.append(review_one_string)
```

```python
len(rev_list)
```

```python
pos_rev = nltk.corpus.movie_reviews.fileids('pos')
```

```python
pos_rev
```

```python
for rev in pos_rev:
    pos_text_neg = rev = nltk.corpus.movie_reviews.words(rev)
    review_one_string= " ".join(rev_text_neg)
    review_one_string = review_one_string.replace(' ,',',')
    review_one_string = review_one_string.replace(' .','.')
    review_one_string = review_one_string.replace("\' " ,"'")
    review_one_string = review_one_string.replace(" \'" ,"'")
    rev_list.append(review_one_string)
```

```python
len(rev_list)
```

```python
neg_targets = np.zeros((1000,),dtype = np.int)
pos_targets = np.ones((1000,),dtype = np.int)
```

```python
target_list = []
for neg_tar in neg_targets:
    target_list.append(neg_tar)
for pos_tar in pos_targets:
    target_list.append(pos_tar)
```

```python
target_list
```

```python
len(target_list)
```

```python
y = pd.Series(target_list)
```

```python
type(y)
```

```python
y.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer(lowercase=True,stop_words='english',min_df=2)
```

```python
x_count_vect = count_vect.fit_transform(rev_list)
x_count_vect
```

```python
x_count_vect.shape
```

```python
x_names = count_vect.get_feature_names()
```

```python
x_names
```

```python
x_count_vect = pd.DataFrame(x_count_vect.toarray(),columns=x_names)
```

```python
x_count_vect.shape
```

```python
x_count_vect.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix
X_train, X_test, y_train, y_test = train_test_split(x_count_vect,y,test_size=0.25,random_state = 5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.naive_bayes import GaussianNB
modelGNB = GaussianNB()
modelGNB_predicted = modelGNB.fit(X_train.toarray(),y_train).predict(X_test.toarray())
modelGNB_predicted
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
modelMNB = MultinomialNB()
modelMNB_predicted = modelMNB.fit(X_train.toarray(),y_train).predict(X_test.toarray())
modelMNB_predicted
```

```python
print(metrics.accuracy_score(y_test,modelGNB_predicted))
```

```python
print(metrics.accuracy_score(y_test,modelMNB_predicted))
```

```python
print(confusion_matrix(y_test,modelGNB_predicted))
```
