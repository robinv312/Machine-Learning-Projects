```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
load_breast_cancer_obj = load_breast_cancer()
```

```python
load_breast_cancer_obj.keys()
```

```python
load_breast_cancer_obj.values()
```

```python
print(load_breast_cancer_obj['DESCR'])
```

REAFDING DATA FROM THE DICTIONARY

```python
features = pd.DataFrame(load_breast_cancer_obj['data'] , columns = load_breast_cancer_obj['feature_names'])
```

```python
features.head()
```

```python
features.info()
```

```python
print(load_breast_cancer_obj['target'])
```

CREATING TARGET AS A DATAFRAME

```python
targetData = pd.DataFrame(load_breast_cancer_obj['target'], columns = ['Cancer'])
```

```python
targetData
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(features,np.ravel(targetData),test_size = 0.30,random_state = 101)
```

```python
from sklearn.svm import SVC
```

```python
svmOject = SVC()
```

```python
svmOject.fit(X_train,y_train)
```

```python
predict = svmOject.predict(X_test)
```

```python
predict
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,predict))
```

GRID SEARCH

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C':[0.1,1,10,100,1000],'gamma':[1,0.1,0.01,0.001,0.0001],'kernel':['rbf']}
```

```python
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_estimator_
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
predict1 = grid.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(classification_report(y_test,predict1))
```
